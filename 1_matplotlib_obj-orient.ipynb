# Free-placing-multi-figures
* start with 'fig', a canvas that can be added axis on
* add_axes: left, bottom, width, height
* axes.plot()
* axes.set_....('..')

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
axes1.plot(x,y)
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('title')
```
