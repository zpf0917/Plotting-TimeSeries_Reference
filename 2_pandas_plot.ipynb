# Correlation heat map

```python
df1_corr = df1.corr()
fig, axes = plt.subplots(figsize=(2,2))
ax = sns.heatmap(df1_corr, vmin = -1, vmax = 1, \
                 xticklabels = df1_corr.columns, yticklabels = df1_corr.columns, \
                 annot = False, ax = axes)
bottom,top = ax.get_ylim()
ax.set_ylim(bottom+0.5, top-0.5)
```
