# Line Plot: format x(time)-axis ticks: plot_date() from matplotlib
* major/ minor locator
* major/ minor formater

```python
import matplotlib.dates as dates
```

```python
idx = mcdon.loc['2007-01-01' : '2007-05-01'].index
stock = mcdon.loc['2007-01-01' : '2007-05-01']['Adj. Close']
```

```python
fig, ax = plt.subplots(figsize = (12,2))
ax.plot_date(idx, stock, fmt = '-')
ax.yaxis.grid(True)
ax.xaxis.grid(True)

# FIRST
ax.xaxis.set_major_locator(dates.MonthLocator())
ax.xaxis.set_major_formatter(dates.DateFormatter('\n\n%b-%Y'))

# SECOND
ax.xaxis.set_minor_locator(dates.WeekdayLocator())
ax.xaxis.set_minor_formatter(dates.DateFormatter('%d'))

# SECOND
# ax.xaxis.set_minor_locator(dates.WeekdayLocator(byweekday = 2))
# ax.xaxis.set_minor_formatter(dates.DateFormatter('%a'))

fig.autofmt_xdate()
```
